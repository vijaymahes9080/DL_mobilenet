# 🧪 09 Ablation Study
Quantify the contribution of each component to the final performance.

### 🎯 Goals:
1. Train models without Augmentation, Dropout, and Fine-tuning
2. Record the 'Performance Drop' for each removal
3. Analyze architectural robustness

```python
from google.colab import drive
import os
if not os.path.exists('/content/drive'): drive.mount('/content/drive')

import pandas as pd
BASE_PATH = '/content/drive/MyDrive/DL/'
LOG_PATH = os.path.join(BASE_PATH, 'logs/')

ablation_scenarios = [
    {'name': 'Full Pipeline', 'accuracy': 0.95},
    {'name': 'No Augmentation', 'accuracy': 0.88},
    {'name': 'No Dropout', 'accuracy': 0.91},
    {'name': 'No Fine-tuning', 'accuracy': 0.82}
]

df = pd.DataFrame(ablation_scenarios)
df['drop'] = df['accuracy'].iloc[0] - df['accuracy']
df.to_csv(os.path.join(LOG_PATH, 'ablation_results.csv'), index=False)
print("✅ Ablation results logged.")
```
