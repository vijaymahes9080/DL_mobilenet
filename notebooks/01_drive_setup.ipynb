# ☁️ 01 Drive Setup
This notebook handles the initial environment setup for the AI Project.

### 🎯 Goals:
1. Mount Google Drive
2. Create the required directory structure
3. Verify GPU availability
4. Install/Update necessary libraries

```python
from google.colab import drive
import os

# 1. Mount Google Drive
drive.mount('/content/drive')

# Define base path
BASE_PATH = '/content/drive/MyDrive/DL/'

# 2. Create Directory Structure
folders = ['dataset', 'processed_data', 'models', 'logs', 'outputs', 'notebooks']

for folder in folders:
    path = os.path.join(BASE_PATH, folder)
    if not os.path.exists(path):
        os.makedirs(path)
        print(f"✅ Created: {path}")
    else:
        print(f"ℹ️ Exists: {path}")
```

### 🏎️ Verify GPU Availability

```python
import tensorflow as tf

gpu_devices = tf.config.list_physical_devices('GPU')
if gpu_devices:
    print(f"✅ GPU Available: {gpu_devices}")
    !nvidia-smi
else:
    print("❌ GPU NOT found. Please go to Runtime > Change runtime type and select GPU.")
```
