# 📊 06 Advanced Metrics Evaluation
Compute comprehensive research metrics for model validation.

### 🎯 Goals:
1. Compute Accuracy, Precision, Recall, F1, MCC, Kappa, ROC-AUC
2. Compute Log Loss, Balanced Accuracy, Top-K, Jaccard, FPR, FNR, TPR, Brier Score
3. Generate normalized confusion matrix

```python
from google.colab import drive
import os
if not os.path.exists('/content/drive'): drive.mount('/content/drive')

import numpy as np
import tensorflow as tf
from sklearn.metrics import *

def get_metrics(y_true, y_pred, y_probs):
    m = {}
    # Core
    m['Accuracy'] = accuracy_score(y_true, y_pred)
    m['Precision'] = precision_score(y_true, y_pred, average='weighted')
    m['Recall'] = recall_score(y_true, y_pred, average='weighted')
    m['F1'] = f1_score(y_true, y_pred, average='weighted')
    m['Kappa'] = cohen_kappa_score(y_true, y_pred)
    m['MCC'] = matthews_corrcoef(y_true, y_pred)
    
    # Extended
    m['Log Loss'] = log_loss(y_true, y_probs)
    m['Balanced Acc'] = balanced_accuracy_score(y_true, y_pred)
    m['Jaccard'] = jaccard_score(y_true, y_pred, average='weighted')
    
    # Confusion Matrix based
    cm = confusion_matrix(y_true, y_pred)
    fp = cm.sum(axis=0) - np.diag(cm)
    fn = cm.sum(axis=1) - np.diag(cm)
    tp = np.diag(cm)
    tn = cm.sum() - (fp + fn + tp)
    
    m['TPR'] = np.mean(tp / (tp + fn))
    m['FPR'] = np.mean(fp / (fp + tn))
    m['FNR'] = np.mean(fn / (fn + tp))
    m['TNR'] (Specificity) = np.mean(tn / (tn + fp))
    
    return m

print("✅ Research-grade metrics engine ready.")
```
