## 🧬 Understanding the Neural Synapse (Deep Learning 101)

ORIEN uses **Convolutional Neural Networks (CNNs)**—specifically the **MobileNetV2** architecture—to perceive information. 

### Why do we use Transfer Learning?
Instead of training from scratch (which takes weeks and millions of images), we use a "pre-trained" brain that already knows how to see basic shapes (lines, curves, textures) from the ImageNet database. We then "fine-tune" only the top layers to recognize your specific modalities like **Gestures** or **Emotions**.

### ⚡ How we make it FASTER:
1. **Resolution Downscaling**: We train at 96x96 pixels. This reduces the number of calculations per layer by 4x compared to HD.
2. **Parallel Prefetching**: While the CPU is training on one batch, the data generator is already loading the next one into RAM.
3. **Layer Freezing**: By locking the first 50 layers, we only calculate gradients for the final decision-making neurons, saving massive compute time.
